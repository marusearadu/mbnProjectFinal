```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
# Make historgrams in 3x2 figure layout

fig, axes = plt.subplots(3, 2, figsize=(12, 12))
axes = axes.flatten()

for i, col in enumerate(data.columns[1:]):
    sns.histplot(data[col], kde=True, ax=axes[i]).set_title(col)

plt.tight_layout()
plt.show()
```
